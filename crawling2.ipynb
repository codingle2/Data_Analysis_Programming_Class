```python

# ==========================================
# 0. í•„ìš” ë¼ì´ë¸ŒëŸ¬ë¦¬
# ==========================================
# Colabì—ëŠ” ëŒ€ë¶€ë¶„ ì„¤ì¹˜ë˜ì–´ ìˆìœ¼ë‚˜, ì—†ìœ¼ë©´ ì£¼ì„ í•´ì œ í›„ ì„¤ì¹˜
# !pip install openpyxl plotly

import os
import time
import requests
import json
import re
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import warnings
warnings.filterwarnings("ignore")

# ê³µí†µ header (ëª¨ë“  ìš”ì²­ì—ì„œ ì‚¬ìš©)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
    "Accept": "application/json"
}

# ==========================================
# 1. ê³µí†µ ìœ í‹¸ í•¨ìˆ˜ / ìƒìˆ˜
# ==========================================
def parse_krw_hangeul(value):
    """
    ê±°ë˜ëŒ€ê¸ˆ(ì–µ) ë‹¨ìœ„ì˜ í•œê¸€ í‘œí˜„ì„ ì •ìˆ˜(ì› ë‹¨ìœ„)ë¡œ ë³€í™˜.
    ì˜ˆ: "1,234ì–µì›" -> 123400000000
    """
    if value is None:
        return 0

    value = str(value).replace(',', '').replace(' ', '')

    if value.endswith('ì–µì›'):
        try:
            num = float(value.replace('ì–µì›', ''))
            return int(num * 100000000)
        except ValueError:
            return 0

    try:
        return int(value)
    except ValueError:
        return 0

# ë„¤ì´ë²„ m.stock ê³µí†µ í—¤ë” (ì°¨ë‹¨ ë°©ì§€ìš©)
HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    ),
    "Referer": "https://m.stock.naver.com/",
    "Origin": "https://m.stock.naver.com",
    "Accept": "application/json, text/plain, */*",
}


def clean_value(value):
    """
    API ê°’ì—ì„œ 'ë°°', 'ì›', 'ì–µì›', ',' ë“±ì„ ì œê±°í•´
    ìˆ«ìë§Œ ë‚¨ê¸°ê±°ë‚˜ ê¹”ë”í•œ ë¬¸ìì—´ë¡œ ì •ë¦¬.
    """
    if value is None or value == "":
        return "N/A"
    return (
        str(value)
        .replace('ë°°', '')
        .replace('ì›', '')
        .replace('ì–µì›', '')
        .replace(',', '')
        .strip()
    )


# ==========================================
# 2. ê±°ë˜ëŸ‰ ìƒìœ„ ì¢…ëª© / KOSPI ì§€ìˆ˜
# ==========================================
def get_top_stocks_data():
    """
    ë„¤ì´ë²„ ê¸ˆìœµ APIì—ì„œ ê±°ë˜ëŸ‰ ìƒìœ„ ì£¼ì‹ ë°ì´í„°ë¥¼ ê°€ì ¸ì™€ DataFrameìœ¼ë¡œ ë°˜í™˜.
    """
    stocks_per_page = 20
    base_api_url = "https://m.stock.naver.com/api/stocks/quantTop/all"

    headers = {
        "User-Agent": (
            "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
            "AppleWebKit/537.36 (KHTML, like Gecko) "
            "Chrome/120.0.0.0 Safari/537.36"
        ),
        "Accept": "application/json, text/plain, */*",
        "Referer": "https://m.stock.naver.com/",
    }

    total_api_pages_to_fetch = 2 # í…ŒìŠ¤íŠ¸ìš©: 2í˜ì´ì§€(40ê°œ). í•„ìš”ì‹œ ëŠ˜ë¦¬ë©´ ë¨.
    fetched_data = []

    print(f"=== ìƒìœ„ ì£¼ì‹ ë°ì´í„° í¬ë¡¤ë§ ì‹œì‘ (ì´ {total_api_pages_to_fetch * stocks_per_page}ê°œ ëª©í‘œ) ===")

    for i in range(1, total_api_pages_to_fetch + 1):
        full_url = f"{base_api_url}?page={i}&pageSize={stocks_per_page}"
        print(f" -> í˜ì´ì§€ {i} ìš”ì²­ ì¤‘...")

        try:
            response = requests.get(full_url, headers=headers, timeout=5)
            response.raise_for_status()
            data = response.json()

            if isinstance(data, dict) and 'stocks' in data and isinstance(data['stocks'], list):
                fetched_data.extend(data['stocks'])
            else:
                print(f" -> í˜ì´ì§€ {i} ë°ì´í„° ì—†ìŒ í˜¹ì€ í˜•ì‹ ì˜¤ë¥˜.")
                break

        except Exception as e:
            print(f" -> í˜ì´ì§€ {i} ì˜¤ë¥˜ ë°œìƒ: {e}")
            break

        time.sleep(0.1) # ì„œë²„ ë¶€í•˜ ë°©ì§€ìš© ë”œë ˆì´

    if not fetched_data:
        return pd.DataFrame()
    # DataFrameìœ¼ë¡œ ë³€í™˜ í›„ ì»¬ëŸ¼ ì •ë¦¬
    df = pd.DataFrame(fetched_data)
    cols = {
        'itemCode': 'ì¢…ëª©ì½”ë“œ',
        'stockName': 'ì¢…ëª©ëª…',
        'closePrice': 'í˜„ì¬ê°€',
        'fluctuationsRatio': 'ë“±ë½ë¥ ',
        'accumulatedTradingVolume': 'ê±°ë˜ëŸ‰',
        'accumulatedTradingValueKrwHangeul': 'ê±°ë˜ëŒ€ê¸ˆ',
    }

    available_cols = [c for c in cols.keys() if c in df.columns]
    df = df[available_cols].rename(columns=cols)

    # ê±°ë˜ëŒ€ê¸ˆ í•­ëª©ì´ ìˆìœ¼ë©´ íŒŒì‹±
    if 'ê±°ë˜ëŒ€ê¸ˆ' in df.columns:
        df['ê±°ë˜ëŒ€ê¸ˆ'] = df['ê±°ë˜ëŒ€ê¸ˆ'].apply(parse_krw_hangeul)

    return df


def get_kospi_info():
    """
    ë„¤ì´ë²„ì—ì„œ KOSPI ê¸°ë³¸ ì •ë³´ ë° ì‹¤ì‹œê°„ ì •ë³´ë¥¼ ê°€ì ¸ì˜¨ë‹¤.
    ë°˜í™˜ê°’: {'basic': ..., 'realtime': ...}
    """
    basic_url = "https://m.stock.naver.com/api/index/KOSPI/basic"
    realtime_url = "https://polling.finance.naver.com/api/realtime/domestic/index/KOSPI"

    result = {}

    # ê¸°ë³¸ ì •ë³´
    try:
        res = requests.get(basic_url, timeout=5).json()
        result['basic'] = res
    except Exception as e:
        print(f"KOSPI ê¸°ë³¸ ì •ë³´ ì˜¤ë¥˜: {e}")
        result['basic'] = None

    # ì‹¤ì‹œê°„ ì •ë³´
    try:
        res = requests.get(realtime_url, timeout=5).json()
        result['realtime'] = res.get('result', res)
    except Exception as e:
        print(f"KOSPI ì‹¤ì‹œê°„ ì •ë³´ ì˜¤ë¥˜: {e}")
        result['realtime'] = None

    return result


def print_kospi_formatted(kospi_data):
    basic = kospi_data.get('basic')
    if not basic:
        print("KOSPI ì •ë³´ë¥¼ ë¶ˆëŸ¬ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")
        return

    # 1) ê¸°ë³¸ ì‹œì„¸ ì •ë³´
    name = basic.get('stockName', 'KOSPI')
    price = basic.get('closePrice', 'N/A')
    change = basic.get('compareToPreviousClosePrice', '0')
    ratio = basic.get('fluctuationsRatio', '0')

    # ë“±ë½ ë°©í–¥
    direction_info = basic.get('compareToPreviousPrice', {})
    direction_text = direction_info.get('text', '')
    symbol = " "
    if direction_text == 'ìƒìŠ¹':
        symbol = "â–²"
    elif direction_text == 'í•˜ë½':
        symbol = "â–¼"

    # ê±°ë˜ì†Œ ì •ë³´
    exchange_info = basic.get('stockExchangeType', {})
    nation = exchange_info.get('nationName', 'KR')
    ex_name = exchange_info.get('nameKor', 'ì½”ìŠ¤í”¼')

    # ì‹œê°„ í¬ë§· ì •ë¦¬
    time_str = basic.get('localTradedAt', '')
    if 'T' in time_str:
        time_str = time_str.split('+')[0].replace('T', ' ')

    # ì¼ë´‰ ì°¨íŠ¸ ì´ë¯¸ì§€ URL
    charts = basic.get('imageCharts', {})
    day_chart = charts.get('day', 'ì •ë³´ ì—†ìŒ')

    # ì¶œë ¥
    print("â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”")
    print(f"â”‚  ğŸ‡°ğŸ‡·  [{nation}] {ex_name} ({name}) ìƒì„¸ í˜„í™©      â”‚")
    print("â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤")
    print(f"â”‚  í˜„ì¬ì§€ìˆ˜  : {price:<33} â”‚")
    print(f"â”‚  ë“±ë½ë¥     : {symbol} {change} ({ratio}%) {direction_text:<15} â”‚")
    print("â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤")
    print(f"â”‚  ì¥ ìƒíƒœ   : {basic.get('marketStatus')}                          â”‚")
    print(f"â”‚  ê¸°ì¤€ì‹œê°„  : {time_str:<26}  â”‚")
    print(f"â”‚  ê±°ë˜ì†Œ    : {exchange_info.get('nameEng')} ({exchange_info.get('code')})              â”‚")
    print("â”œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”¤")
    print("â”‚  ğŸ“Š ì¼ë´‰ ì°¨íŠ¸ (í´ë¦­í•˜ì—¬ ë³´ê¸°):                   â”‚")
    print(f"â”‚  {day_chart:<46}  â”‚")
    print("â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜")


# ==========================================
# 3. ì¢…ëª©ì½”ë“œ / ì‹œì„¸ / ì¬ë¬´ / í†µí•© ì •ë³´
# ==========================================
def get_code_from_excel(stock_name, file_name='stockCodeInfo.xlsx'):
    """
    ë¡œì»¬ ì—‘ì…€ íŒŒì¼ì—ì„œ 'íšŒì‚¬ëª…'ìœ¼ë¡œ ì¢…ëª©ì½”ë“œë¥¼ ì°¾ëŠ”ë‹¤.
    - ì¢…ëª©ì½”ë“œëŠ” 0ìœ¼ë¡œ ì‹œì‘í•˜ë¯€ë¡œ ë¬¸ìì—´(str)ë¡œ ì½ì–´ì•¼ í•œë‹¤.
    """
    if not os.path.exists(file_name):
        print(f"ì˜¤ë¥˜: '{file_name}' íŒŒì¼ì´ í˜„ì¬ í´ë”ì— ì—†ìŠµë‹ˆë‹¤.")
        # ì—…ë¡œë“œí•œ íŒŒì¼ ì¤‘ ë™ì¼ ì´ë¦„ì´ ìˆëŠ”ì§€ ì‹œë„
        if file_name in uploaded.keys():
            # Colab ì—…ë¡œë“œ ì‹œ íŒŒì¼ì€ í˜„ì¬ ê²½ë¡œì— ì €ì¥ë¨
            print(f" -> ì—…ë¡œë“œëœ íŒŒì¼ì´ ìˆìœ¼ë¯€ë¡œ ê³„ì† ì§„í–‰í•©ë‹ˆë‹¤.")
        else:
            return None

    try:
        df = pd.read_excel(file_name, dtype={'ì¢…ëª©ì½”ë“œ': str})
        found_row = df[df['íšŒì‚¬ëª…'] == stock_name]

        if not found_row.empty:
            return str(found_row.iloc[0]['ì¢…ëª©ì½”ë“œ'])

        print(f"'{stock_name}'ì„(ë¥¼) ì—‘ì…€ íŒŒì¼ì—ì„œ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None

    except Exception as e:
        print(f"ì—‘ì…€ íŒŒì¼ ì½ê¸° ì˜¤ë¥˜: {e}")
        return None


def get_daily_stock_prices(code, pages=1):
    """
    ì¢…ëª©ì½”ë“œ(code)ë¥¼ ë°›ì•„ ì¼ë³„ ì‹œì„¸ë¥¼ ê°€ì ¸ì™€ DataFrameìœ¼ë¡œ ë°˜í™˜.
    í˜¸ì¶œ URL: https://m.stock.naver.com/api/stock/{code}/price
    """
    url = f"https://m.stock.naver.com/api/stock/{code}/price"

    all_prices = []

    headers = {
        "User-Agent": "Mozilla/5.0",
        "Referer": "https://m.stock.naver.com/"
        }

    print(f"=== [{code}] ì‹œì„¸ ë°ì´í„° ìš”ì²­ ì¤‘... ===")

    for page in range(1, pages + 1):
        params = {"pageSize": 20, "page": page}

        try:
            response = requests.get(url, params=params, headers=headers, timeout=5)
            data = response.json()
            if len(data) > 0:
                all_prices.extend(data)
                print(f" -> {page}í˜ì´ì§€ ìˆ˜ì§‘ ì„±ê³µ ({len(data)}ê±´)")
            else:
                break

        except Exception as e:
            print(f"API ìš”ì²­ ì˜¤ë¥˜: {e}")
            break

        time.sleep(0.1) # ì°¨ë‹¨ ë°©ì§€ìš© ë”œë ˆì´

    if not all_prices:
        return pd.DataFrame()

    df = pd.DataFrame(all_prices)

    # ì»¬ëŸ¼ëª… í•œê¸€í™”
    rename_map = {
        'localTradedAt': 'ë‚ ì§œ',
        'closePrice': 'ì¢…ê°€',
        'compareToPreviousClosePrice': 'ì „ì¼ë¹„',
        'openPrice': 'ì‹œê°€',
        'highPrice': 'ê³ ê°€',
        'lowPrice': 'ì €ê°€',
        'accumulatedTradingVolume': 'ê±°ë˜ëŸ‰',
        'fluctuationsRatio': 'ë“±ë½ë¥ ', # ìˆìœ¼ë©´ í•¨ê»˜ ì‚¬ìš©
    }

    available = [c for c in rename_map.keys() if c in df.columns]
    return df[available].rename(columns=rename_map)


def get_financial_summary(code):
    """
    ì—°ê°„ ì¬ë¬´ ìš”ì•½ ì •ë³´ ê°€ì ¸ì˜¤ê¸°.
    í˜¸ì¶œ URL: https://m.stock.naver.com/api/stock/{code}/finance/annual
    ë°˜í™˜: í–‰=í•­ëª©, ì—´=ì—°ë„(YYYYMM) ì¸ DataFrame
    """
    url = f"https://m.stock.naver.com/api/stock/{code}/finance/annual"
    res = requests.get(url, headers=headers)

    if res.status_code != 200:
        print(" ì¬ë¬´ì œí‘œ API ì˜¤ë¥˜")
        return pd.DataFrame()

    data = res.json()
    finance = data.get("finance", None)

    #  ì¶”ê°€í•´ì•¼ í•  ì˜ˆì™¸ ì²˜ë¦¬ ë¶€ë¶„
    if finance is None:
        print("âš  ì¬ë¬´ì œí‘œ ë°ì´í„°ê°€ ì œê³µë˜ì§€ ì•ŠëŠ” ì¢…ëª©ì…ë‹ˆë‹¤.")
        return pd.DataFrame()

    titles = finance.get("trTitleList", [])
    years = [t["key"] for t in titles]

    trList = finance.get("trList", [])
    rows = []
    for item in trList:
        label = item["title"]
        values = item["trValue"]
        row = {"ì§€í‘œ": label}
        row.update({year: values[i] for i, year in enumerate(years)})
        rows.append(row)

    df = pd.DataFrame(rows)
    return df


def get_stock_overview(code):
    """
    í†µí•©(integration) ì—”ë“œí¬ì¸íŠ¸ë¥¼ í™œìš©í•´
    ì¬ë¬´/ìˆ˜ê¸‰/ë¦¬í¬íŠ¸/ì‚°ì—…ë¹„êµ ë“± ë‹¤ì–‘í•œ ì •ë³´ë¥¼ í•œ ë²ˆì— ê°€ì ¸ì˜¨ë‹¤.

    í˜¸ì¶œ URL: https://m.stock.naver.com/api/stock/{code}/integration

    ë°˜í™˜:
        {
            "totalInfos": {...},
            "dealTrend": DataFrame,
            "researchReports": list,
            "industryCompare": DataFrame,
            "consensus": {...},
            "corporationSummary": {...}
        }
    """
    url = f"https://m.stock.naver.com/api/stock/{code}/integration"

    try:
        res = requests.get(url, headers=HEADERS, timeout=5)
        data = res.json()
    except Exception:
        return {}

    result = {}

    # â‘  í•µì‹¬ ê¸°ì—…ì§€í‘œ (ì‹œì´, PER, 52ì£¼ ìµœê³ /ìµœì € ë“±)
    total_infos = data.get("totalInfos", [])
    total_info_dict = {}
    for item in total_infos:
        key = item.get("key")
        value = item.get("value")
        total_info_dict[key] = value
    result["totalInfos"] = total_info_dict

    # â‘¡ ìµœê·¼ 5ì¼ ìˆ˜ê¸‰
    deals = data.get("dealTrendInfos", [])
    deal_list = []
    for d in deals:
        deal_list.append({
            "bizdate": d.get("bizdate"),
            "closePrice": d.get("closePrice"),
            "compareToPreviousClosePrice": d.get("compareToPreviousClosePrice"),
            "foreigner": d.get("foreignerPureBuyQuant"),
            "institution": d.get("organPureBuyQuant"),
            "individual": d.get("individualPureBuyQuant"),
            "volume": d.get("accumulatedTradingVolume"),
        })
    result["dealTrend"] = pd.DataFrame(deal_list)

    # â‘¢ ì¦ê¶Œì‚¬ ë¦¬í¬íŠ¸
    result["researchReports"] = data.get("researches", [])

    # â‘£ ì‚°ì—… ë‚´ ê²½ìŸì‚¬ ë¹„êµ
    industry_list = []
    industry_raw = data.get("industryCompareInfo", [])
    for info in industry_raw:
        industry_list.append({
            "ì¢…ëª©ëª…": info.get("stockName"),
            "ì¢…ê°€": info.get("closePrice"),
            "ë“±ë½": info.get("compareToPreviousClosePrice"),
            "ë“±ë½ë¥ ": info.get("fluctuationsRatio"),
            "ì‹œì´": info.get("marketValue"),
        })
    result["industryCompare"] = pd.DataFrame(industry_list)

    # â‘¤ ëª©í‘œê°€ ì»¨ì„¼ì„œìŠ¤
    result["consensus"] = data.get("consensusInfo", {})

    # â‘¥ íšŒì‚¬ ê°œìš”
    result["corporationSummary"] = data.get("corporationSummary", {})

    return result


def print_comprehensive_report(stock_name, stock_code):
    """
    í•œ ì¢…ëª©ì— ëŒ€í•´:
    - í•µì‹¬ ì§€í‘œ
    - ìµœê·¼ ìˆ˜ê¸‰
    - ì¦ê¶Œì‚¬ ë¦¬í¬íŠ¸
    - ëª©í‘œì£¼ê°€ ì»¨ì„¼ì„œìŠ¤
    - ê²½ìŸì‚¬ ë¹„êµ
    - ì—°ê°„ ì¬ë¬´ì œí‘œ
    ë¥¼ ì½˜ì†”ì— ì¢…í•© ì¶œë ¥.
    """
    if not stock_code:
        return

    print("\n" + "=" * 60)
    print(f" ğŸ“¢ [{stock_name}] ({stock_code}) ì¢…í•© ë¶„ì„ ë¦¬í¬íŠ¸")
    print("=" * 60 + "\n")

    overview = get_stock_overview(stock_code)
    financial = get_financial_summary(stock_code)

    # 1) í•µì‹¬ì§€í‘œ ìš”ì•½
    info = overview.get("totalInfos", {})
    important_keys = [
        ("ì‹œê°€ì´ì•¡", "ì‹œì´"),
        ("52ì£¼ ìµœê³ ", "52ì£¼ ìµœê³ "),
        ("52ì£¼ ìµœì €", "52ì£¼ ìµœì €"),
        ("PER", "PER"),
        ("EPS", "EPS"),
        ("ì¶”ì •PER", "ì¶”ì •PER"),
        ("ì¶”ì •EPS", "ì¶”ì •EPS"),
        ("PBR", "PBR"),
        ("BPS", "BPS"),
        ("ë°°ë‹¹ìˆ˜ìµë¥ ", "ë°°ë‹¹ìˆ˜ìµë¥ "),
        ("ì£¼ë‹¹ë°°ë‹¹ê¸ˆ", "ì£¼ë‹¹ë°°ë‹¹ê¸ˆ"),
    ]

    print("[1] í•µì‹¬ ì§€í‘œ ìš”ì•½")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    for label, key in important_keys:
        value = info.get(key)
        if value not in ["", None, "-"]:
            print(f"{label:<12} : {value}")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€\n")

    # 2) ìµœê·¼ ìˆ˜ê¸‰
    print("[2] ìµœê·¼ 5ì¼ íˆ¬ìì ìˆœë§¤ìˆ˜")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    deal_df = overview.get("dealTrend")
    if isinstance(deal_df, pd.DataFrame) and not deal_df.empty:
        print(deal_df)
    else:
        print("ë°ì´í„° ì—†ìŒ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€\n")

    print("[3] ìµœì‹  ì¦ê¶Œì‚¬ ë¦¬í¬íŠ¸ (ìµœê·¼ 5ê°œ)")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")

    # 3) ì¦ê¶Œì‚¬ ë¦¬í¬íŠ¸
    reports = overview.get("researchReports", [])
    if reports:
        for r in reports[:5]:
            title = r.get("tit", "")
            broker = r.get("bnm", "")
            date = r.get("wdt", "")
            print(f"- {broker} : \"{title}\" ({date})")
    else:
        print("ë°ì´í„° ì—†ìŒ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€\n")

    # 4) ëª©í‘œì£¼ê°€ ì»¨ì„¼ì„œìŠ¤
    print("[4] ëª©í‘œì£¼ê°€ ì»¨ì„¼ì„œìŠ¤")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    consensus = overview.get("consensus", {})
    if consensus:
        print(f"í‰ê·  ëª©í‘œê°€ : {consensus.get('priceTargetMean')}ì›")
        print(f"í‰ê·  ì˜ê²¬   : {consensus.get('recommMean')}")
        print(f"ì‘ì„±ì¼     : {consensus.get('createDate')}")
    else:
        print("ë°ì´í„° ì—†ìŒ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€\n")

    # 5) ê²½ìŸì‚¬ ë¹„êµ
    print("[5] ê²½ìŸì‚¬ ë¹„êµ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    ind_df = overview.get("industryCompare")
    if isinstance(ind_df, pd.DataFrame) and not ind_df.empty:
        print(ind_df)
    else:
        print("ë°ì´í„° ì—†ìŒ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€\n")

    # 6) ì—°ê°„ ì¬ë¬´ì œí‘œ
    print("[6] ì—°ê°„ ì¬ë¬´ì œí‘œ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")
    if isinstance(financial, pd.DataFrame) and not financial.empty:
        print(financial)
    else:
        print("ë°ì´í„° ì—†ìŒ")
    print("â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€")

    print("\n" + "=" * 60 + "\n")


# ==========================================
# 4. ì¶”ê°€: ì „ì²˜ë¦¬/í”¼ì²˜ì—”ì§€ë‹ˆì–´ë§/EDA í•¨ìˆ˜ (ì¶”ê°€ëœ ë¶€ë¶„)
# ==========================================
def process_time_series(df_ts, date_col='ë‚ ì§œ', price_col='ì¢…ê°€', volume_col='ê±°ë˜ëŸ‰', ticker=None):
    """
    ë‹¨ì¼ ì¢…ëª©ì˜ ì¼ë³„ ì‹œê³„ì—´ DataFrameì„ ë°›ì•„ì„œ:
    - ë‚ ì§œ ì •ë¦¬, ê²°ì¸¡ ì²˜ë¦¬, ì´ìƒì¹˜ ì²˜ë¦¬(ìŒìˆ˜/0 ì œê±°)
    - ì´ë™í‰ê· (MA5, MA20, MA60), ê±°ë˜ëŸ‰ ë³€í™”ìœ¨, ìˆ˜ìµë¥ (1/5/20), ë³€ë™ì„±(20ì¼), Drawdown
    - íƒ€ê¹ƒ(ë‹¤ìŒë‚  ìˆ˜ìµë¥ , 5ì¼ ìˆ˜ìµë¥ ) ìƒì„±
    ë°˜í™˜: ì²˜ë¦¬ëœ DataFrame
    """
    df = df_ts.copy()

    # ë‚ ì§œ ì»¬ëŸ¼ ëª…í™•í™” ë° ì •ë ¬
    if date_col in df.columns:
        df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
        df = df.sort_values(by=date_col).reset_index(drop=True)
    else:
        # ë§Œì•½ ë‚ ì§œ ì»¬ëŸ¼ì´ ì—†ìœ¼ë©´ ì¸ë±ìŠ¤ë¥¼ ë‚ ì§œë¡œ ê°„ì£¼í•˜ì§€ ì•ŠìŒ
        pass

    # ìˆ«ìí˜• ë³€í™˜: ì‰¼í‘œ ì œê±° í›„ numeric ë³€í™˜ (ì•ˆì •ì„±)
    for c in [price_col, volume_col]:
        if c in df.columns:
            df[c] = df[c].astype(str).str.replace(',', '').replace('nan', np.nan)
            df[c] = pd.to_numeric(df[c], errors='coerce')

    # ê²°ì¸¡ì¹˜ ì²˜ë¦¬: ì„ í˜• ë³´ê°„ -> ì• ì±„ì›€ -> ë’¤ ì±„ì›€ -> ë‚¨ìœ¼ë©´ ì œê±°
    df = df.sort_values(by=date_col) if date_col in df.columns else df
    if price_col in df.columns:
        df[price_col] = df[price_col].interpolate(method='linear')
        df[price_col] = df[price_col].fillna(method='ffill').fillna(method='bfill')

    if volume_col in df.columns:
        df[volume_col] = df[volume_col].fillna(0)

    # ì´ìƒì¹˜ ì²˜ë¦¬: ìŒìˆ˜/0 ê°€ê²© ì œê±° (0ì¸ ê²½ìš°ëŠ” ê±°ë˜ì •ì§€ ë“± ê°€ëŠ¥ì„± ìˆìœ¼ë‹ˆ ë³´ì¡´ ì—¬ë¶€ëŠ” ì‚¬ìš©ìê°€ ì„ íƒ)
    if price_col in df.columns:
        df = df[df[price_col] > 0].copy()

    # ì´ë™í‰ê· 
    for w in [5, 20, 60]:
        if price_col in df.columns:
            df[f"MA{w}"] = df[price_col].rolling(window=w, min_periods=1).mean()

    # ê±°ë˜ëŸ‰ ë³€í™”ìœ¨
    if volume_col in df.columns:
        df['vol_change'] = df[volume_col].pct_change().replace([np.inf, -np.inf], np.nan)

    # ìˆ˜ìµë¥  (ì¼ê°„/5ì¼/20ì¼)
    if price_col in df.columns:
        df['ret_1d'] = df[price_col].pct_change(1)
        df['ret_5d'] = df[price_col].pct_change(5)
        df['ret_20d'] = df[price_col].pct_change(20)

    # ë³€ë™ì„±(20ì¼ í‘œì¤€í¸ì°¨)
    if 'ret_1d' in df.columns:
        df['volatility_20d'] = df['ret_1d'].rolling(20).std()

    # Drawdown: í˜„ì¬ê°€ / ìµœê·¼ 60ì¼ ìµœê³  - 1
    if price_col in df.columns:
        roll_max = df[price_col].rolling(window=60, min_periods=1).max()
        df['drawdown'] = df[price_col] / roll_max - 1

    # íƒ€ê¹ƒ ìƒì„±: ë‹¤ìŒë‚  ìˆ˜ìµë¥ , 5ì¼ ìˆ˜ìµë¥ 
    if 'ret_1d' in df.columns:
        df['target_ret_1d'] = df['ret_1d'].shift(-1)
    if 'ret_5d' in df.columns:
        df['target_ret_5d'] = df['ret_5d'].shift(-5)

    # ì¢…ëª©ëª…/í‹°ì»¤ ì»¬ëŸ¼ ì‚½ì… (ìˆìœ¼ë©´ ìœ ì§€, ì—†ìœ¼ë©´ ì¶”ê°€)
    if ticker:
        df['ticker'] = ticker
    else:
        if 'ì¢…ëª©ëª…' not in df.columns and 'ticker' not in df.columns:
            df['ticker'] = 'UNKNOWN'

    # ìµœì¢… ì •ë ¬
    if date_col in df.columns:
        df = df.sort_values(by=date_col).reset_index(drop=True)

    return df


def eda_and_visualize(df, date_col='ë‚ ì§œ', price_col='ì¢…ê°€', save_fig=False, prefix='eda'):
    """
    ê¸°ë³¸ì ì¸ EDA ì‹œê°í™”:
    - ê°€ê²© + MA (plotly)
    - ìˆ˜ìµë¥  ë¶„í¬ (hist), ë°•ìŠ¤í”Œë¡¯
    - ìˆ˜ì¹˜í˜• ìƒê´€ê´€ê³„ íˆíŠ¸ë§µ (ìˆ«ìí˜• ì»¬ëŸ¼ë§Œ ì‚¬ìš©)
    ê²°ê³¼: None (í™”ë©´ ì¶œë ¥), í•„ìš”í•˜ë©´ ì €ì¥ ê°€ëŠ¥
    """
    # ì•ˆì „: ìˆ«ì ì»¬ëŸ¼ë§Œ ê³¨ë¼ì„œ ìƒê´€ê³„ìˆ˜ ê³„ì‚°
    numeric_df = df.select_dtypes(include=[np.number])

    # 1) Plotly: ê°€ê²© & ì´ë™í‰ê· 
    try:
        if price_col in df.columns:
            fig = go.Figure()
            fig.add_trace(go.Scatter(x=df[date_col] if date_col in df.columns else df.index,
                                     y=df[price_col], mode='lines', name='Close'))
            for ma in ['MA5', 'MA20', 'MA60']:
                if ma in df.columns:
                    fig.add_trace(go.Scatter(
                        x=df[date_col] if date_col in df.columns else df.index,
                        y=df[ma], mode='lines', name=ma))
            fig.update_layout(title='Price & Moving Averages', xaxis_title='Date', yaxis_title='Price')
            fig.show()
    except Exception as e:
        print("Plotly ê°€ê²© ê·¸ë˜í”„ ìƒì„± ì¤‘ ì˜¤ë¥˜:", e)

    # 2) ìˆ˜ìµë¥  ë¶„í¬, ë°•ìŠ¤í”Œë¡¯
    try:
        if 'ret_1d' in df.columns:
            plt.figure(figsize=(10,4))
            sns.histplot(df['ret_1d'].dropna(), bins=100, kde=False)
            plt.title('1-day Return Distribution')
            plt.show()

            plt.figure(figsize=(8,4))
            sns.boxplot(x=df['ret_1d'].dropna())
            plt.title('1-day Return Boxplot')
            plt.show()
    except Exception as e:
        print("ìˆ˜ìµë¥  ë¶„í¬/ë°•ìŠ¤í”Œë¡¯ ìƒì„± ì¤‘ ì˜¤ë¥˜:", e)

    # 3) ìƒê´€ê´€ê³„ íˆíŠ¸ë§µ (ìˆ«ìí˜•ë§Œ)
    try:
        if numeric_df.shape[1] > 1:
            plt.figure(figsize=(10,8))
            sns.heatmap(numeric_df.corr(), cmap='coolwarm', center=0)
            plt.title('Correlation Heatmap (numeric only)')
            plt.show()
        else:
            print("Heatmap ìƒëµ â€” ìˆ«ìí˜• ì»¬ëŸ¼ì´ ë¶€ì¡±í•©ë‹ˆë‹¤.")
    except Exception as e:
        print("íˆíŠ¸ë§µ ìƒì„± ì¤‘ ì˜¤ë¥˜:", e)

def plot_extra_eda(df):
    import matplotlib.pyplot as plt
    import seaborn as sns

    # ===== ìˆ«ìí˜• ë³€í™˜ ì²˜ë¦¬ (ì‰¼í‘œ ì œê±° í›„ float ë¡œ ë³€í™˜) =====
    df_numeric = df.copy()
    for col in df_numeric.columns:
        if df_numeric[col].dtype == 'object':
            df_numeric[col] = df_numeric[col].astype(str).str.replace(',', '', regex=False)
            df_numeric[col] = pd.to_numeric(df_numeric[col], errors='ignore')

    # íˆìŠ¤í† ê·¸ë¨ (ì¼ê°„ ìˆ˜ìµë¥ )
    if 'ì¼ê°„ìˆ˜ìµë¥ ' in df_numeric.columns:
        plt.figure(figsize=(8, 4))
        df_numeric['ì¼ê°„ìˆ˜ìµë¥ '].dropna().hist(bins=50)
        plt.title("ì¼ê°„ ìˆ˜ìµë¥  ë¶„í¬")
        plt.xlabel("ìˆ˜ìµë¥ ")
        plt.ylabel("ë¹ˆë„")
        plt.show()

    # ë°•ìŠ¤í”Œë¡¯ (ê°€ê²© / ê±°ë˜ëŸ‰)
    for cols in [['ì¢…ê°€', 'ê±°ë˜ëŸ‰'], ['MA5', 'MA20', 'MA60']]:
        valid_cols = [c for c in cols if c in df_numeric.columns]
        if len(valid_cols) > 0:
            plt.figure(figsize=(8, 4))
            df_numeric[valid_cols].plot.box()
            plt.title(f"{', '.join(valid_cols)} ë°•ìŠ¤í”Œë¡¯")
            plt.show()

    # ìƒê´€ê´€ê³„ íˆíŠ¸ë§µ
    plt.figure(figsize=(10, 8))
    corr = df_numeric.corr(numeric_only=True)
    sns.heatmap(corr, cmap='coolwarm', annot=False)
    plt.title("ìƒê´€ê´€ê³„ Heatmap")
    plt.show()


# ==========================================
# 5. ì‹¤í–‰ ì˜ˆì‹œ ë° í†µí•© íŒŒì´í”„ë¼ì¸ (ê¸°ì¡´ main ìœ ì§€ + ì¶”ê°€ ì²˜ë¦¬)
# ==========================================
if __name__ == "__main__":
    # 1) ê±°ë˜ëŸ‰ ìƒìœ„ ì¢…ëª© ëª©ë¡
    print("\n[1] ê±°ë˜ëŸ‰ ìƒìœ„ ì£¼ì‹ ëª©ë¡")
    df_top = get_top_stocks_data()
    if not df_top.empty:
        # ì£¼ìš” ì»¬ëŸ¼ë§Œ ë³´ê¸°
        display_cols = [c for c in ['ì¢…ëª©ëª…', 'í˜„ì¬ê°€', 'ë“±ë½ë¥ ', 'ê±°ë˜ëŸ‰'] if c in df_top.columns]
        print(df_top[display_cols].head(5))
    else:
        print("ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")

    # 2) KOSPI ì •ë³´
    print("\n[2] KOSPI ì •ë³´")
    kospi_data = get_kospi_info()
    print_kospi_formatted(kospi_data)

    # 3) íŠ¹ì • ì¢…ëª© ì‹œì„¸ + ì¢…í•© ë¦¬í¬íŠ¸
    target_name = "ì‚¼ì„±ì „ì"
    stock_code = get_code_from_excel(target_name)

    if stock_code:
        print(f"\n[3] '{target_name}' ì‹œì„¸ ì¡°íšŒ (ì½”ë“œ: {stock_code})")
        df_result = get_daily_stock_prices(stock_code, pages=5)  # í•„ìš”í•˜ë©´ í˜ì´ì§€ ëŠ˜ë¦¬ê¸°

        if not df_result.empty:
            print("\n[ìµœì‹  ë°ì´í„° 5ê±´]")
            if 'ë“±ë½ë¥ ' in df_result.columns:
                display(df_result[['ë‚ ì§œ', 'ì¢…ê°€', 'ì „ì¼ë¹„', 'ë“±ë½ë¥ ', 'ê±°ë˜ëŸ‰']].head(5))
            else:
                display(df_result.head(5))

            # ì¢…í•© ë¦¬í¬íŠ¸ ì¶œë ¥
            print_comprehensive_report(target_name, stock_code)

            # =========================
            # ì¶”ê°€: ì „ì²˜ë¦¬ + EDA íŒŒì´í”„ë¼ì¸
            # =========================
            print("\n[4] ì „ì²˜ë¦¬ Â· í”¼ì²˜ì—”ì§€ë‹ˆì–´ë§ Â· EDA ìˆ˜í–‰ ì¤‘...")

            # 1) ì‹œê³„ì—´ ì²˜ë¦¬
            processed = process_time_series(df_result, date_col='ë‚ ì§œ', price_col='ì¢…ê°€', volume_col='ê±°ë˜ëŸ‰', ticker=target_name)

            # 2) ê¸°ë³¸ EDA ë° ì‹œê°í™”
            eda_and_visualize(processed, date_col='ë‚ ì§œ', price_col='ì¢…ê°€')

            # 3) ìµœì¢… ì €ì¥ (í´ë” ìƒì„± ë° CSV ì €ì¥)
            out_dir = "data/processed"
            os.makedirs(out_dir, exist_ok=True)
            out_path = os.path.join(out_dir, f"merged_data_{stock_code}.csv")
            processed.to_csv(out_path, index=False)
            print(f"âœ… ì „ì²˜ë¦¬ ë°ì´í„° ì €ì¥ ì™„ë£Œ: {out_path}")

            # 4) ê°„ë‹¨í•œ EDA ìš”ì•½ ìë™ ì¶œë ¥ (ì˜ˆì‹œ â€” í•„ìš”ì‹œ í™•ì¥)
            print("\n[5] ê°„ë‹¨í•œ EDA ìš”ì•½")
            # ì˜ˆ: ë³€ë™ì„±(í‰ê· ), ìµœê·¼ 20ì¼ ìˆ˜ìµë¥  í‰ê· , ìµœëŒ€ ë‚™í­
            mean_vol = processed['volatility_20d'].mean() if 'volatility_20d' in processed.columns else np.nan
            recent_ret_20d = processed['ret_20d'].tail(20).mean() if 'ret_20d' in processed.columns else np.nan
            max_dd = processed['drawdown'].min() if 'drawdown' in processed.columns else np.nan

            print(f"- í‰ê·  20ì¼ ë³€ë™ì„±: {mean_vol:.6f}")
            print(f"- ìµœê·¼ 20ì¼ í‰ê·  20ì¼ ìˆ˜ìµë¥ (í‰ê· ): {recent_ret_20d:.6f}")
            print(f"- ìµœëŒ€ ë‚™í­ (drawdown, ìŒìˆ˜ê°’): {max_dd:.6f}")

        else:
            print("ì‹œì„¸ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")
    else:
        print("\n[3] ì¢…ëª© ì½”ë“œë¥¼ í™•ë³´í•˜ì§€ ëª»í•´ ì „ì²˜ë¦¬ íŒŒì´í”„ë¼ì¸ì€ ì‹¤í–‰ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

if stock_code:

    df_result = get_daily_stock_prices(stock_code)

    if not df_result.empty:
        # ì¢…í•© ë¦¬í¬íŠ¸ ì¶œë ¥
        print("\n[ ì¢…í•© ë¦¬í¬íŠ¸]")
        print_comprehensive_report(target_name, stock_code)

        # ===== ì‹œê³„ì—´ ì²˜ë¦¬ & EDA ë¶„ì„ =====
        print("\n[ ì‹œê³„ì—´ ë°ì´í„° ì „ì²˜ë¦¬/FE ì‹œì‘]")
        df_processed = process_time_series(
            df_result,
            date_col='ë‚ ì§œ',
            price_col='ì¢…ê°€',
            volume_col='ê±°ë˜ëŸ‰',
            ticker=target_name
        )
        print("ì²˜ë¦¬ ì™„ë£Œ â€” ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°:")
        display(df_processed.tail())

        print("\n[ EDA ì‹œê°í™” ì‹¤í–‰]")
        eda_and_visualize(df_processed, date_col='ë‚ ì§œ', price_col='ì¢…ê°€')
        print(" EDA ì™„ë£Œ")

    else:
        # ì´ elseëŠ” 'if not df_result.empty'ì— ëŒ€ì‘
        print("ì¼ë³„ ì‹œì„¸ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")

else:
    # ì´ elseëŠ” 'if stock_code'ì— ëŒ€ì‘
    print("ì¢…ëª©ì½”ë“œë¥¼ ì°¾ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")


print("\n[ ì‹œê³„ì—´ ë°ì´í„° ì „ì²˜ë¦¬/FE ì‹œì‘]")
df_processed = process_time_series(
    df_result,
    date_col='ë‚ ì§œ',
    price_col='ì¢…ê°€',
    volume_col='ê±°ë˜ëŸ‰',
    ticker=target_name
)

# ê²°ì¸¡ì¹˜ ì²˜ë¦¬
df_processed = df_processed.sort_values('ë‚ ì§œ')
df_processed['ì¢…ê°€'] = df_processed['ì¢…ê°€'].interpolate().ffill()
df_processed['ê±°ë˜ëŸ‰'] = df_processed['ê±°ë˜ëŸ‰'].fillna(0)

# ì´ìƒì¹˜ ì²˜ë¦¬
df_processed = df_processed[df_processed['ì¢…ê°€'] > 0]
df_processed = df_processed[df_processed['ê±°ë˜ëŸ‰'] >= 0]

# í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§ ì¶”ê°€
df_processed['MA5'] = df_processed['ì¢…ê°€'].rolling(5).mean()
df_processed['MA20'] = df_processed['ì¢…ê°€'].rolling(20).mean()
df_processed['MA60'] = df_processed['ì¢…ê°€'].rolling(60).mean()

df_processed['ì¼ê°„ìˆ˜ìµë¥ '] = df_processed['ì¢…ê°€'].pct_change()
df_processed['ì£¼ê°„ìˆ˜ìµë¥ '] = df_processed['ì¢…ê°€'].pct_change(5)
df_processed['ì›”ê°„ìˆ˜ìµë¥ '] = df_processed['ì¢…ê°€'].pct_change(20)

df_processed['ë³€ë™ì„±_20'] = df_processed['ì¼ê°„ìˆ˜ìµë¥ '].rolling(20).std()
df_processed['ê³ ì ëŒ€ë¹„í•˜ë½ë¥ '] = 1 - (df_processed['ì¢…ê°€'] / df_processed['ì¢…ê°€'].cummax())

df_processed['ê±°ë˜ëŸ‰ë³€í™”ìœ¨'] = df_processed['ê±°ë˜ëŸ‰'].pct_change()

print("ì²˜ë¦¬ ì™„ë£Œ â€” ë°ì´í„° ë¯¸ë¦¬ë³´ê¸°:")
display(df_processed.tail())

# CSV ì €ì¥
df_processed.to_csv("data/processed/merged_data.csv", index=False)
print(" ì €ì¥ ì™„ë£Œ â†’ data/processed/merged_data.csv")

# í™•ì¥ëœ EDA ì‹¤í–‰
print("\n[ EDA ì‹œê°í™” ì‹¤í–‰]")
eda_and_visualize(df_processed, date_col='ë‚ ì§œ', price_col='ì¢…ê°€')
plot_extra_eda(df_processed)   # ìƒˆë¡œìš´ ì‹œê°í™” í•¨ìˆ˜ ì¶”ê°€
print(" EDA ì™„ë£Œ")

# ê±°ë˜ëŸ‰ ìƒìœ„ 1ìœ„ ì¢…ëª©
if __name__ == "__main__":
    top_df = get_top_stocks_data()
    first_stock = top_df.iloc[0]
    stock_name = first_stock['ì¢…ëª©ëª…']
    stock_code = first_stock['ì¢…ëª©ì½”ë“œ']

    print(f"\n ì˜¤ëŠ˜ ê±°ë˜ëŸ‰ 1ìœ„ ì¢…ëª© ìë™ ì„ ì •: {stock_name} ({stock_code})")

    print_comprehensive_report(stock_name, stock_code)

    df_price = get_daily_stock_prices(stock_code, pages=10)
    df_processed = process_time_series(df_price, ticker=stock_name)
    eda_and_visualize(df_processed)

# ==========================================
# 6. ì‹¤í–‰ íŒŒì´í”„ë¼ì¸
# ==========================================
stock = input("ì¡°íšŒí•  ì¢…ëª©ëª…ì„ ì…ë ¥í•˜ì„¸ìš”: ")

code = get_code_from_excel(stock)   # ì—‘ì…€ì—ì„œ ì¢…ëª©ì½”ë“œ ì°¾ê¸°
if not code:
    raise ValueError(" ì¢…ëª©ì½”ë“œë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")

# (1) ì¼ë³„ ì‹œì„¸ ê°€ì ¸ì˜¤ê¸°
ts = get_daily_stock_prices(code, pages=10)   # pages = 10 â†’ ìµœê·¼ 200ì¼
if ts.empty:
    raise ValueError(" ì‹œì„¸ ë°ì´í„°ë¥¼ ë¶ˆëŸ¬ì˜¤ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")

# (2) ì „ì²˜ë¦¬ + í”¼ì²˜ ì—”ì§€ë‹ˆì–´ë§
ts_processed = process_time_series(ts, ticker=stock)

# (3) ê¸°ë³¸ EDA ì‹œê°í™”
eda_and_visualize(ts_processed)

# (4) ì¢…í•© ë¦¬í¬íŠ¸ ì¶œë ¥ (ì¬ë¬´/ìˆ˜ê¸‰/ë¦¬í¬íŠ¸/ê²½ìŸì‚¬ ë“±)
print_comprehensive_report(stock, code)

print("\n ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤.")




```
